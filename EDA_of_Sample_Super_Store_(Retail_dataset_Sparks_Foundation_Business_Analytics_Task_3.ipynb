```python
sales_state = a.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Sales')
```

```python
sns.countplot(x='Sub-Category',data=a)
figsize=(12,7)
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
a.groupby('Sub-Category')['Profit'].sum().nlargest(10).sort_values(ascending=False).plot(kind='pie', ax=ax1)
plt.xlabel('Product_ID')
```
